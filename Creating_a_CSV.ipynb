# **DATE RANGE**

```python
#yy-mm-dd - days frequency
pd.date_range(start= '2024-04-01', end='2024-04-12')
```

```python
#months frequency
pd.date_range(start= '2024-04-01', end='2024-09-12', freq = 'M')
```

```python
#Weekdays frequency
pd.date_range(start='2024-02-01',end='2024-02-10',freq='B')
```

# **TOP COMMANDS**



*   **creating new, renaming and droping columns**
*   **Apply method, Sort index, Sort values**




```python
data
```

```python
#Creating New columns
data['city'] = ['Delhi', 'bangalore', 'Chennai', 'Patna', 'Pune', 'Pune', 'Mumbai']
```

```python
data
```

```python
#Renaming Columns
data.columns = ['NAME','AGE','UNIVERSITY','CITY']
```

```python
data
```

```python
#Droping Columns
data.drop(['AGE'], axis=1)    #temporary_dropping
```

```python
data
```

```python
data.drop(['AGE'], axis=1, inplace=True)           #permanent dropping
```

```python
data
```

```python
#APPLY_METHOD
data['MARKS'] = [201,202,195,180,201,190,225]
data
```

```python
data['MARKS'] = data['MARKS'].apply(lambda x: x/5)          #using lambda fxn
data
```

```python
data.columns = ['NAME','UNIVERSITY','CITY', 'MARKS/SUBJECT']
data
```

```python
#SORT_INDEX
data.index = [10,28,64,7,1,6,15]                #assigning Index
data
```

```python
data.sort_index()           #ascending sorting of index : default value
```

```python
data.sort_index(ascending=False)     #descending sorting of index
```

```python
data.sort_values(by='MARKS/SUBJECT')   #SORTING BY VALUES: ASC ORDER: DEFAULT ORDER
```

```python
data.sort_values(by='MARKS/SUBJECT', ascending=False)
```
