```python
population = np.arange(5,50,5)         ###Population variance
population
```

```python
pop_mean = sum(population)/len(population)
```

```python
pop_mean
```

```python
sq_pop_deviation = [(i-pop_mean)**2 for i in population]
sq_pop_deviation
```

```python
pop_variance = sum(sq_pop_deviation)/len(population)
```

```python
print('Population Variance : {}'.format(round(pop_variance)))
```

```python
sample = np.arange(5,25,5)      #Sample variance
sample
```

```python
s_mean = sum(sample)/len(sample)
s_mean
```

```python
sq_s_deviation = [(i-s_mean)**2 for i in sample]
sq_s_deviation
```

```python
s_variance = sum(sq_s_deviation)/(len(sample)-1)
```

```python
print('Sample variance: {}'.format(round(s_variance, 1)))
```

#  *NOTE:* **Lesser the value of variance the more accurate it is**

# **Standard Deviation**

```python
import math
```

```python
std_dev = math.sqrt(pop_variance)
```

```python
print('Standard Deviation : {}'.format(round(std_dev,1)))
```
