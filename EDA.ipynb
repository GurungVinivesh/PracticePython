#**Handling Outliers**

```python
#Handling outliers using IQR(Interquartile range)
list_ = [1,2,3,4,5,6,9,10,12,55,66]
```

```python
Q1 = np.percentile(list_,25)
Q1
```

```python
Q3 = np.percentile(list_,75)
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
#now to identify the lower quartile and upper quartile
low = Q1 - 1.5 * IQR
up = Q3 + 1.5 * IQR
low, up
```

```python
Outliers = []
```

```python
for i in list_:
  if((i>up)or(i<low)):
    Outliers.append(i)
```

```python
Outliers
```

```python
for i in list_:
  if i not in Outliers:
    print(i)
```

```python
list_ = [i for i in list_ if i not in Outliers]
list_
```
